## Investigation Overview


>In this investigation, I wanted to look at the features of bike sharing system that could be used to analyze rides and trips around the bay area . The main focus was on duration of trips, age and customer type.



## Dataset Overview

>  This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area. The Ford Go Bike Data can be downloaded from here: https://www.fordgobike.com/system-data

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
color = sns.color_palette()[0]

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_clean = pd.read_csv('clean_ford_gobike.csv')
```

## Distribution Usage by Hour

> Usage by hour shows the start hour users start their ride. Plotting usage by hour shows that users mostly use the system in the morning hours of 8am and 9am and in the afternoon hours of 5pm and 6pm. 


```python
plt.figure(figsize=[10,6])

sns.countplot(data=df_clean, x='start_hour', color=color)
plt.title('Usage by Hour', fontsize=15, fontweight='bold')
plt.xlabel('Hours', fontweight='bold')
plt.ylabel('Number of Bike Rides', fontweight='bold');
```

## Distribution Usage by hour between User Type

> There are two types of users: Customer and Subscriber. The plot shows that Subscribers uses the system more than customers. It seems like subscribers are working class people that uses the system to commute to work while Customers are tourists and other people that uses the system for fun to explore the bay area

```python
plt.figure(figsize=[10,6])
sns.countplot(data=df_clean, x='start_hour', hue='user_type');
plt.xlabel('Hours', fontweight='bold')
plt.ylabel('Number of Bike Rides', fontweight='bold');
```

## Users Age Distribution

> The age of users in the dataset was created from the member year column. The age is plotted to show the distribution age of the users. The age is between of 20-66 years. Most of the users age are between 20-40 years. 



```python
plt.figure(figsize=[10,6])
bins = np.arange(0, 65, 5)

plt.hist(data = df_clean, x = 'age', bins = bins);
plt.xlim([18,61])
plt.title("User by Age", y=1.05, fontweight='bold')
plt.xlabel('Age of Users', fontweight='bold')
plt.ylabel('Number of Bike Trips', fontweight='bold');
```

## User Age Distribution between User Type
> We have seen the age distribution of the users, we explore more of the age distribution between user type to see the distribution in Customers and subscribers. Subscribers have wider range of ages while customers tends to be younger.

```python
plt.figure(figsize=[10,6])

plot = sns.catplot(data = df_clean.query('age <= 70'), x = 'age', col='user_type', kind='count', sharey = False, color = color, height=5, aspect=2);
plot.set_axis_labels('User Age', 'Number of Bike Trips');
plot.set_titles('{col_name}');
plot.fig.suptitle('Customers vs Subscribers Ages', fontsize=16, fontweight='bold');
plot.set_xticklabels(rotation=80);
```
